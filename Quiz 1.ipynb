```python
import pandas as pd
import numpy as np
import matplotlib as mpl
```

# ცხრილის შენახვა ცვლადში

```python
df = pd.read_csv(r"survey_results_public.csv")
```

# ცხრილის სასურველი სტრიქონების გამოტანა და ინდექსირება

```python
selected_data = df.loc[1:40, ['Age', 'WorkExp', 'RemoteWork']]
selected_data
```

# ორი ფილტრი სამუშაო გამოცდილების და მუშაობის რეჟიმის მიხედვით

```python
workExpStrRemover = pd.to_numeric(selected_data['WorkExp'], errors='coerce')
workExpFilter = workExpStrRemover < 20
remoteFilter = selected_data['RemoteWork'] == 'Remote'

filt = selected_data[workExpFilter & remoteFilter]
filt
```

# სორტირება

```python
sortedList = filt.sort_values(by=['WorkExp', 'Age'], ascending=[True, True])
sortedList
```

# მოდა, მედიანა, საშუალო, სტანდარტული გადახრა, მინ, მაქს

```python
mediana = sortedList['WorkExp'].median()
moda = sortedList['WorkExp'].mode()
mean = sortedList['WorkExp'].mean()
std = sortedList['WorkExp'].std()
minimum = sortedList['WorkExp'].min()
maximum = sortedList['WorkExp'].max()

print('mediana = ', mediana)
print('moda = ', moda.to_list())
print('sashualo = ', mean)
print('standartuli gadaxra = ', std)
print('min = ', minimum)
print('max = ', maximum)

```

```python
import matplotlib.pyplot as plt
```

# Bar Graph 

```python
plt.bar(sortedList['Age'], sortedList['WorkExp'] )
plt.xlabel('Age')
plt.ylabel('Work Experience (Years)')
```
